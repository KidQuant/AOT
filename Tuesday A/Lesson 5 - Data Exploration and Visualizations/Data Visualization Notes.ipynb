## Line Charts

Here we will take a first look at creating a simple line plot. As with everything we're going to be doing, we will import all of the important ploting modules.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# plt.style.use('seaborn-v0_8-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

So now we've created our first chart line chart in python. 

If we want, we can create multiple lines on the same plot by simply calling the plot function multiple times.

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

### Adjusting Line Colors and Styles

So now we have a line chart for Sine and Cosine. 

Great! 

That's all there is to plotting a simple line chart in Matplotlib. 

But maybe you want something a little more to line chart. Maybe you want to change the colors; maybe you want to make the lines dashed or dotted. 

Python gives you those options, which we will see in the next few examples.

The `plt.plot()` method takes additional arugments that can be used to adjust colors. 

To adjust the color, just use the keyword `color` and the color that you want.

```python
# Keep this code chunk

plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

You can specify the color by name, by rgb, grayscale, whatever. 

Those of you who do graphic design might know about all of these things, but the points is you're not limited in the amount of parameters you can change.

Let's just say you want to change the style of the line. 

All you have to do is use the keyword `linestyle` in the `plt.plot()` method.

```python
# Keep this code chunk

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');
```

### Changing Axes

Matplotlib does a decent job of allowing you to customize your axis of your plots.

Take the sine plot from before

```python
# keep this chunk
plt.plot(x, np.sin(x))
```

Notice how there is a little bit extra at the end beyond the 0 and 10. 

We can fix this with two methods `xlim()` and `ylim()`. 

Let's copy the code from the top except this time, we'll add a few things

```python
plt.plot(x, np.sin(x))

#add this after copying top code
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
```

So now our chart looks little bit better. 

There's lots of ways we can adjust this axis. 

You can refer to the matplotlib documentation for more information on that.

### Labeling Plots

So now that we have a nicer looking plot, lets include labels to make it more official. 

Let's copy the code from above.

```python
plt.plot(x, np.sin(x))
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# add this after copying code from above
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('f(x)')
```

So now we have an official chart that we have just created on python. 

It has a title, it has the axis all correctly labeled. 

It has everything you could possibly want for a simple line plot, and most of all, we've done it in just a few lines of code.

Let's do one more example with a legend, because sometimes you want to be able to identify what you're looking at. 

So I'll copy the code from above except we're going to add a few things

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# copy the code from above and change parameters

plt.title('Trig Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Bar Chart

```python
# Keep this code chunk

data = {'C':20, 'C++':15, 'Java':30, 
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon', 
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

## Scatter Plots

So for this example we're going to use Iris dataset from Scikit-Learn python library so let's import that. 

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

So before we create a scatter plot from this data, let's figure out exactly what it is we are working with.

To do this, we need the help of the pandas library, so let's import pandas.

```python
import pandas as pd
```

Now that we have pandas imported, we can create a dataframe from the available data.

```python
df = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
df
```

So as you can see from the DataFrame, we're working with 4 types of data: Sepal length, sepal width, petal length and petal width.

Let's make a scatter plot from sepal length and sepal width.

**Alpha**: Alpha is a parameter that makes the points more or less transparent

**s**: S is for the size of each point. I want the size of each point to correspond with the petal width, so we're going to use the 3rd feature and scale it up for a bit.

**C**: C allows us to set the color of each point. So there is three types of flowers in the Iris data set Setosa, Versicolour, and Virginica

```python
plt.scatter(features[0], features[1], 
            alpha=.2, 
            s=features[3]*100, 
            c=iris.target, 
            cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

### The Average

In order to illustrate this example, I'm going to need the help of the random library. So I'm going to import random, which is a built in library for Python.

```python
import random
```

So I'm gonna create a random sample of 11 integers from 12  to 30

```python
int_list=[]

for i in range(11):
    int_list.append(random.randint(3,12))
print(int_list)
```

As you can see, I have 11 random integers.

If I want to calculate the average of this list, I would simply added up all of the values and divide by the number of values. 

So I would added up all 11 values and then divide by 11.

```python
np.sum(int_list)/len(int_list)
```

So we use this to get the average of a list of numbers all with one line of code.

Similarly, we could have just used the `np.average()` method from Numpy, but I actually wanted you to understand, for those of you who are still lost, on how to find the average of a data set.

```python
np.average(int_list)
```

We do this and we calculate the average just fine.

## Histogram

First we're going to create a simple histogram. 

We already have the infastructure in place to do this, so all we need is one simple line of code.

We're going to use `np.random.randn()` function. 

```python
data = np.random.randn(1000)
```

This will return a random sample of numbers from a standard normal distribution. 

So essential we are replicating a standard normal distribution. 

Now I've said standard normal distribution, when in the slides I have just normal distribution. 

Is there a difference? Yes. 

Standard Normal Distribution is a distribution where the mean is 0 and the standard deviation is 1. 

That's all you really need to know. I won't get into the specifics.

Now we're going to plot using the `plt.hist()` methed.

```python
plt.hist(data, edgecolor="black")
```

As we can see, pretty symmetrical bell curve. 


One peak, and the tails are about the same. 

Just about what you would expect from a standard normal distribution.

We can customize this if we want, by taking the code as before but changing a few things.

*start coding*

**Bins**: Explain what bins are.

**histtype**: histtype allows you to set how you want the histogram to be drawn.

**color**: color should be self explanatory


```python
plt.hist(data, bins=25, alpha=0.5,
histtype='stepfilled', color='steelblue');
```

So now we have histogram that is more representative of the number of observations in our sample, AND is stylistically more appealing.

We can use this same stylistic choice to show an overlap between two different distributions.

The following code shows an example of this.

The code represents 3 different distributions and we're using the normal distribution instead of the standard normal distribution.

`x1` has a mean of 0 and standard deviatioin of 0.8, `x2` has a mean of -2 and a standard deviation of 1, and `x3` has a mean of 3 and a standard deviation of 2.

I've created a throw away function that applies the same style to all three histograms.

Let's run this code

```python
# keep this code chunk

x1 = np.random.normal(0, 0.6, 1000)
x2 = np.random.normal(-3, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=20)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

As we can see the blue histogram has very steep peek, because the standard deviation of that one is very small.

The green histogram is relatively flat because that one has a higher standard deviation.

So now we've seen an enough examples of normal distributions.

Let's look at distributions that may not be normal, and for this we're going to use a dataset from the `seaborn` library.

So lets import seaborn.

```python
import seaborn as sns
```

Now let's load the tips dataset from the seaborn library by doing the following:

```python
tips = sns.load_dataset('tips')
tips
```

Notice how seaborn already gives us a DataFrame. 

We don't need to do any data curation or anything. It's already available to us.

So this dataframe has 7 columns: `total_bill`, `tips`, `sex`, `smoker`, `day`, `time`,and `size`. 

We can conduct some descriptive statistics of our own.

First we lets calculate average bill. 

We could do this with the `np.average()` but Pandas already has a functionality that allows us to do this. 

To calculate the average we just simply write

```python
tips['total_bill'].mean()
```

So the average bill in our dataset is roughly 19.78. How about the median?

```python
tips['total_bill'].median()
```

So write off the bat we see that the mean is greater than the median. 

It looks like this distribution could be right skewed. 

But how can we be sure? We can find the standard deviation, which is simple enough.

```python
tips['total_bill'].std()
```

Very high standard deivation.

So what's going on? It looks like there's a lot of variability. 

So at this point, we're going to create a historgram to see what's going on.

```python
plt.hist(tips['total_bill'], bins=15, edgecolor='black')
```

As we can see, our supicions of the distribution being right skewed was correct.

As we can see, the tail is longer on the right than it is on the left. 

The outlier appears to be somewhere on the 15th bin, or somwhere around 50 dollars. 

We can find this with another Pandas method, called `.max()`

```python
tips['total_bill'].max()
```

So the largest bill is 50.81.

Just for kicks, we can find the smallest bill by using the `.min()` method.

```python
tips['total_bill'].min()
```
