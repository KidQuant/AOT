# Data Exploration and Visualizations


## Line Charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### Adjusting Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');
```

### Changing Axes

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.title('Trig Functions')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Bar Chart

```python
# dictionary object for the data
data = {'C':20, 'C++':15, 'Java':30, 
        'Python':35} 

# converts the keys and values to list objects 
courses = list(data.keys())
values = list(data.values())

# changes the dimenions of the chart
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon', 
        width = 0.9)
 
# labels and titles
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

## Pie Chart

```python

# a list of labels and values
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

## Scatter Plots

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
df
```

```python
plt.scatter(features[0], features[1],
           alpha = .2,
           s=features[3] *100,
            c=iris.target,
           cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

## Histogram

```python
data = np.random.randn(1000)
plt.hist(data, edgecolor='black')
```

```python
plt.hist(data, bins=25, alpha=0.5, histtype='stepfilled', color='steelblue')
```

```python
x1 = np.random.normal(0, 0.6, 1000)
x2 = np.random.normal(-3, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=20)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```
