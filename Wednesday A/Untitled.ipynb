```python
import matplotlib.pyplot as plt # imported matplotlib
import numpy as np # imported the numpy package
from sklearn.datasets import load_iris # imported the sklearn 
```

```python
 # assigns the iris data to a variable called 'iris'
iris = load_iris()

# assigns random number generator to a variable called 'rng'
rng = np.random.RandomState(3)

# assigns random floating point numbers to variables called 'x', 'y', and 'colors'
x, y, colors = rng.randn(3, 100)

# annother assignment using random varibles
size = 1000 * rng.rand(100)

plt.scatter(x, y, c= colors, s=size, alpha=0.3, cmap ='viridis')
```
