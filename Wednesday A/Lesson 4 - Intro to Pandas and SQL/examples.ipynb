## Appending Datasets with Python

```python
## As a data scientist you may receive data in multiple datasets.
## This helps with transferring data in smaller chunks instead of
## one large dataset.

## Load in your datasets
df1 = pd.read_csv('C:/Users/user/folder/data/filename1.csv')
df2 = pd.read_csv('C:/Users/user/folder/data/filename2.csv')
```

```python
# Append the datasets together
# Append will add to the bottom
df_append = df1.append(df2)
df_append.describe(include='all')
```

```python
df_append.head(25)
```

```python
# .dropna will remove all missing values inplace = True means to drop the entire row where data is missing
df_append.dropna(inplace=True)
df_append.head(25)
```

```python
### Droping based on a value
df_append.drop(df_append.loc[df_append['spend_2021']==3647.50].index, inplace=True)
df_append.head(25)
```

```python
### Create a dummy attribute for a specific column
df_dc = pd.get_dummies(df_append, columns=['gender'])
df_dc.head()
```
