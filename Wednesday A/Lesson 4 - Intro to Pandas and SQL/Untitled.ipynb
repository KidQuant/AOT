```python
df1 = pd.read_csv("data curation_assignment_data_1.csv")
df2 = pd.read_csv("data curation_assignment_data_2.csv")
```

```python
df3 = pd.concat([df1, df2])
df3.head()
```

```python
sqldf(
'''
select first_name, last_name, email, loyalty_status
from df3
''')
```

```python
sqldf(
"""
select first_name, last_name, email, loyalty_status
from df3
where loyalty_status == 'Silver' 
"""

)
```

```python
sqldf(
'''
SeleCt first_name aS 'First Name', last_name As 'Last Name', email, loyalty_status
from df3
WHERE total_spend_2021 < 3000
limit 5
''')
```

```python
sqldf(
'''
select loyalty_status AS 'Loyalty Status', education AS 'Education', total_spend_2021 AS 'Total Spend'
from df3
where most_traveled_state == 'NY'
''')
```
